# Multiple Variable Analysis

```python
sns.countplot(data['Gender'], hue=(data['Product']))
                                   #Here I wanted to plot the diffence in product user spilt by Gender,Two Categorical variables
                                  #We can see the split between product usage on gender is very similar between the TM1, and TM2
                                                                    #There is a big difference beween usage by Gender of the TM3
```

```python
figure = plt.figure(figsize=(20,10))

df = sns.regplot(x='Treadmill Usage',y='Miles', data=data )
# This plots a scatterplot with a regression line showing a positive correlation between Usage and Miles
```

```python
plt.figure(figsize=(15,5))

sns.pointplot(x="Treadmill Usage", y="Miles", hue = 'Product',  data=data)
plt.show()

                                                                                 #here we are plotting Treadmill Usage and Miles
                                     # We can see that the TM1 (blue) and TM2(orange) models have roughly similar usage patterns
                               # The TM3 has the highest overall usage and a higher mileage to usage ratio than the other models
```

```python
plt.figure(figsize=(15,5))

sns.pointplot(x="Treadmill Usage", y="Fitness Level", hue = 'Product',  data=data)
plt.show()

                                        #this is the same graph but with fitness level indicated against Treadmill Usage as well
                                        #The TM3 model Fitness level generally starts and remains above that of the other Models
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x=data.Product, y=data['Treadmill Usage'])
plt.show()
                                                                    #showing the distribtions of  Treadmill Usage across Models
                    # Again we see the TM1, and TM2 models exhibiting similar usage overlap, the TM3 model has the highest usage
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=data,x='Gender',y='Salary',hue='Product')
plt.show()

                                                                # This bar plot shows the Product usage separted by income level
                         # Shows the TM1 and TM2 are owned by users typically below or just above the mean income level (53,000)
                                                               # the TM3 shows typical owners with an above average income level
```

```python
sns.boxplot(data['Salary'], data['Product']);
                                                        # This boxplot again shows the Salary spread among the Product variables
                                              # Appears that the TM1 and TM2 essentially encompass a very similar income bracket
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=data,x='Product',y='Salary')
plt.show()
                                                      # This just shows the same data without separating by the Gender variable.
                             # Across the board the TM1 has owners with the lowest income, TM2 owners have slightly higer income
                                           # TM3 typically has owners with much higher income levels with a median around 75,000
```
